```python
import cv2
cap=cv2.VideoCapture(0)
if not cap.isOpened():
    print('webcam openening error')
    exit()
while cap.isOpened():
    rti,frame=cap.read()
    if not rti:
        print('frames error')
        break
    results=model.predict(frame)
    frame_width=results[0].plot()
    cv2.imshow('webcam segmentation',frame_width)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
